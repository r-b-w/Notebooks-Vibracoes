Vamos admitir que a massa total da viga é $M$, o comprimento total é $l$, e que serão usadas $N$ massas concentradas, igualmente espaçadas, para representar a viga.

Vamos tomar como coordenada generalizada o deslocamento lateral de cada massa, chamado $y_i$. O valor de cada massa concentrada será igual à massa de um intervalo, pois vamos somar meio intervalo de cada lado da massa.

Para facilitar nossa vida, vamos resolver este problema usando a matriz de flexibilidade, que é trivialmente calculada neste problema, caso saibamos a fórmula para a deflexão de uma viga com uma carga em posição arbitrária, o que podemos encontrar com muita facilidade.

Por exemplo, do apêndice B do Rao temos

$$ y(x) =
\begin{cases}
{\displaystyle\frac{Pbx}{6EIl}}(l^2 - x^2 - b^2)& 0 \le x \le a \\
{\displaystyle\frac{Pa(l-x)}{6EIl}}(2lx - x^2 - a^2)& a \le x \le l,
\end{cases}
$$

onde $a$ é a distância do ponto de aplicação da força ao apoio esquerdo e $b = l-a$.

## Dados do problema

Vamos tratar um problema com uma viga de seção circular vazada, isto é, um tubo, com comprimento igual a 2 metros, diâmetro externo igual a 30 milímetros, espessura de parede igual a 1 milímetro e feita de aço, com módulo de Young igual a 210GPa e massa específica igual a 7700 kg/m³.

Para uma barra com outro tipo de seção transversal é necessário alterar o cálculo do momento de inércia de área abaixo.

```python
import numpy as np

l = 2.0
ϕe = 0.030
t = 0.001
ρ = 7700
E = 210.0e9

ϕi = ϕe - 2.0*t
A = np.pi*(ϕe**2 - ϕi**2)/4.0
vol = A*l
M = vol*ρ
print(f'Total mass of the bar is {M} kg.')
I = np.pi*(ϕe**4 - ϕi**4)/64.0
print(f'Area moment of inertia is {I} m⁴.')
```

## Número de graus de liberdade
O número de graus de liberdade é o número de massas concentradas, e o número de segmentos é um a mais do que o número de graus de liberdade.

```python
NGL = 21
```

## Matriz de massa

```python
nseg = NGL + 1
mseg = M/nseg       # massa concentrada de um seguimento.
m = np.zeros((NGL, NGL), dtype=np.float64)
np.fill_diagonal(m, mseg)
```

## Matriz de flexibilidade

Neste problema é claro que é mais fácil calcular a matriz de flexilibilidade, devido ao esquema mostrado acima.

Como a matriz é simétrica, podemos calcular apenas a parte triangular superior e replicá-la. 

É claro que para um problema genérico esta matriz precisa ser calculada da forma apropriada.

```python
def deflection(P, a, l, E, I, x):
    y = P*a*(l-x)/(6.*E*I*l)*(2.0*l*x - x**2 - a**2)
    return y
```

```python
lseg = l/nseg
x = np.linspace(lseg, l-lseg, NGL)
print(f"Posições das massas concentradas:\n{x[:6]}")
```

```python
a = np.empty((NGL, NGL),  dtype=np.float64)

for row in np.arange(0, NGL):
    a[row,row:] = deflection(1.0, x[row], l, E, I, x[row:])
    a[row,:row] = a[row, :NGL-row-1:-1]
```

## Frequências naturais e modos normais.

Vamos calcular as frequências naturais e modos normais como  um problema de autovalores da matriz dinâmica $\boldsymbol{D} = \boldsymbol{k}^{-1}\boldsymbol{m} = \boldsymbol{a}\boldsymbol{m}$.

O problema a ser resolvido é então
$$ \lambda \boldsymbol{X} = \boldsymbol{D} \boldsymbol{X}.$$

onde $\lambda = 1/\omega^2$. 

Para problemas de grande porte, existem algoritmos que calculas apenas os primeiro $r$ autovalores e autovetores. Infelizmente não temos estes algoritmos diretamente implementados no NumPy, mas há uma função no SciPy que faz isto.

Esta funcão é um pouco mais complicada e faz mais coisas que as rotinas do NumPy, então é necessário usar mais parâmetros e configurações.

Como queremos as frequências naturais mais baixas, vamos calcular os autovalores mais altos!

Temos que definir o número de autovalores e vetore a serem calculados, que tem que ser no máximo $NGL -1 $.


```python
from scipy.sparse.linalg import eigs
```

```python
r = 4       # número de autovalores
D = a@m

v0 = np.zeros(NGL+2, dtype=np.float64) # isto não é necessário, mas é fácil neste caso
v0[1:-1] = x*x[::-1]

λ, X = eigs(D, r, which='LM', v0=v0[1:-1]) 
x.shape
```
